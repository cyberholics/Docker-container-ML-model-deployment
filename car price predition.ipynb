```python
import pandas as pd
import numpy as np
from sklearn.model_selection import  train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_log_error
```

```python
car_data=pd.read_csv("train.csv")
car_data.head()
```

```python
car_data
```

## Data cleaning

```python
#removing the 'km' from the mileage column and converting it to float
car_data['Mileage'] = pd.to_numeric(car_data.Mileage.str.split(' ').str[0], downcast='float')
```

```python
#replacing all the '0' values with the mean values of the 'Mileage' column
car_data['Mileage'] = np.where(car_data['Mileage'] == 0.0,car_data['Mileage'].mean(),car_data['Mileage'])
```

```python
#cleaning the Doors column
car_data['Doors'] = np.where((car_data['Doors'] == '04-May') | (car_data['Doors'] == '02-Mar'), car_data['Doors'].str.split('-').str[0],car_data['Doors'])
```

```python
#Remove the word 'Turbo' from all records that have it
car_data['Engine volume'] = pd.to_numeric(car_data['Engine volume'].str.split(' ').str[0], downcast='float')
```

```python
#Replacing the '0' in the 'Engine volume' column with the mean value of that column
car_data['Engine volume']=np.where(car_data['Engine volume'] == 0.0,car_data['Engine volume'].mean(),car_data['Engine volume'])
```

```python
#Feature engineering the production year column
import datetime as dt
currt_time = dt.datetime.now()
car_data['Prod. year'] = currt_time.year - car_data['Prod. year'] 
```

```python
#converting the Levy column to float as it is the Tax 
car_data['Levy'] = pd.to_numeric(car_data['Levy'].replace('-', '0'), downcast='float')
#Replacing the 0 in the 'Levy' column with mean of that column
car_data['Levy'] = np.where(car_data['Levy'] == 0.0,car_data['Levy'].mean(),car_data['Levy'])
```

```python
car_data
```

### preparing the data for modeling

```python
categ = car_data.select_dtypes(include='object')
num = car_data.select_dtypes(include = np.number)
#getting dummies for the categorical variables
cat_dummies = pd.get_dummies(categ,drop_first=True)
#creating the final dataset
final_car_data = pd.concat([num,cat_dummies], axis=1)
#checking the shape of the final dataset
```

```python
final_car_data
```

### building the model

```python
X = final_car_data.drop(["Price"],axis=1)
Y= final_car_data.Price
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=10)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
```

```python
RMLSE1=np.sqrt(mean_squared_log_error(np.exp(y_test),np.exp(pred)))
```
